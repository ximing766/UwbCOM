```python
import pyecharts.options as opts
from pyecharts.charts import Line,Boxplot,Scatter3D
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv(".\..\..\data\Corr_content.csv", sep=',', header=0, usecols=[0,1,2])
data.head(5)
```

```python
scatter_data = [list(z) for z in zip(data['x'], data['y'], data['z'])]
```

```python
fig = plt.figure()

# 添加一个子图，并指定其为3D图
ax = fig.add_subplot(111, projection='3d')

# 绘制散点图，这里使用 scatter_data 中的 x, y, z 值
ax.scatter([p[0] for p in scatter_data], [p[1] for p in scatter_data], [p[2] for p in scatter_data])

# 设置坐标轴标签
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# 显示图形
plt.show()
```
